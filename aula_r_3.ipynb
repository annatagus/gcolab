Gráfico de Barras

```r
ggplot(vendas, aes(x=Produto, y=Vendas)) +
  geom_bar(stat="identity", fill = "violet")
```

Gráfico de Linhas

```r
ggplot(faturamento, aes(x=Mês, y=Faturamento, group=1)) +
  geom_line (color= "purple", linewidth=3) +
  geom_point (color="orange", size=5) +
  # geom_bar(stat="identity", fill = "violet") +
  geom_text (aes(label = Faturamento), size=6) +
  labs (title="natal?", x="mês", y="faturação em dólares")
```

Gráfico de Pizza

```r
ggplot(categorias, aes(x="",y = Quantidade,fill= Categoria)) +
  geom_col(width =1) +
  coord_polar(theta = "y")
```

de Dispersão
