```r
Df_Housing <- read.csv("/content/sample_data/california_housing_train.csv")
```

```r
head(Df_Housing, 10)
```

```r
summary(Df_Housing)
```

```r
dim(Df_Housing)
```

Seleccionar as colunas que queremos
SELECT()

```r
Df_Housing %>%
           select(median_income, median_house_value)
```

Filtar a tabela
FILTER()

```r
Df_Housing %>% filter(median_income >= 4)
```

Filtrar utilizando E

```r
Df_Housing %>% filter(median_income >= 4, total_rooms < 1000)
```

Ranking de valor de casa

```r
Df_Housing %>%  arrange(desc(median_house_value))
```

```r
Df_Housing %>%  arrange(median_house_value)
```

```r
Df_Housing %>%  arrange(desc(median_house_value)) %>% head(10)
```

```r
top10_porHouseValue <- Df_Housing %>%  arrange(desc(median_house_value)) %>% head(10)
```

```r
summary(top10_porHouseValue)
```

Criar novas Colunas
-> MUTATE()

```r
top10_porHouseValue %>%  mutate( novacoluna = 1)
```

```r
top10_porHouseValue %>%  mutate( Valor_porQuartos = median_house_value / total_bedrooms)
```

```r
Top10_porHouseValueComColunaNova <- top10_porHouseValue %>%  mutate( Valor_porQuartos  =  median_house_value / total_bedrooms)
```

```r
summary(Top10_porHouseValueComColunaNova)
```
